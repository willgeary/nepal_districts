# District and Cluster Assessment Reponses

```python
responses = pd.read_excel("District and Cluster Assessment (Responses).xlsx")
```

```python
responses.head(2)
```

```python
responses.shape
```

```python
fname = "./npl_admbnd/npl_admbnda_districts_nd_20181210.json"

districts = gpd.read_file(fname)
```

```python
districts.head()
```

```python
districts.shape
```

```python
districts.plot()
```

```python
districts.head()
```

```python
unique_districts = list(districts['DIST_EN'].unique())
```

```python
'Bhojpur' in unique_districts
```

```python
unique_districts = {}

for i, row in responses.iterrows():
    org_name = str(row['Organization Name']).rstrip()
    unique_districts[org_name] = []
    
    try:
        districts_prov1 = row['PROVINCE 1: [Province 1 (Koshi)]'].split(", ")
        for j in districts_prov1:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    
    try:
        districts_prov2 = row['PROVINCE 2: [Province 2 (Janaki Pradesh)]'].split(", ")
        for j in districts_prov2:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    try:
        districts_prov3 = row['PROVINCE 3: [Province 3 (Bagmati)]'].split(", ")
        for j in districts_prov3:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    try:
        districts_prov4 = row['PROVINCE 4: [Province 4 (Gandaki Pradesh)]'].split(", ")
        for j in districts_prov4:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    try:
        districts_prov5 = row['PROVINCE 5: [Province 5 (Rapti)]'].split(", ")
        for j in districts_prov5:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    try:
        districts_prov6 = row['PROVINCE 6: [Province 6 (Karnali Pradesh)]'].split(", ")
        for j in districts_prov6:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
    
    try:
        districts_prov7 = row['PROVINCE 7 [Province 7 (Sudurpashchim Pradesh)]'].split(", ")
        for j in districts_prov7:
            if j not in unique_districts[org_name]:
                unique_districts[org_name].append(j)
    except:
        pass
```

```python
unique_districts['Actionaid']
```

```python
output = pd.DataFrame()
organizations = list(unique_districts.keys())
output['organization'] = organizations
district_names = list(unique_districts.values())
output['districts'] = district_names
output = output.sort_values(by='organization', ascending=True)
output.reset_index(drop=True, inplace=True)
```

```python
output.head()
```

```python
districts.head(2)
```

```python
orgs_in_district = {}
for i, row in districts.iterrows():
    district = row['DIST_EN']
    orgs_in_district[district] = []
    for j, org in output.iterrows():
        organization = org['organization']
        this_org_districts = list(org['districts'])
        if district in this_org_districts:
            orgs_in_district[district].append(organization)
```

```python
orgs_in_district_df = pd.DataFrame()
unqiue_districts = list(orgs_in_district.keys())
orgs_in_district_df['DIST_EN'] = unqiue_districts
this_orgs = list(orgs_in_district.values())
orgs_in_district_df['organizations'] = this_orgs
```

```python
orgs_in_district_df.head()
```

```python
districts.head()
```

```python
final = pd.merge(districts, orgs_in_district_df, how='left', on='DIST_EN')
```

```python
final['organizations'] = final['organizations'].apply(lambda x: str(x).replace("[","").replace("]",""))
final = final[['DIST_EN', 'ADM1_PCODE', 'organizations', 'geometry']]
```

```python
final.columns = ['District', 'Province', 'Orgs', 'geometry']
```

```python
final.to_file("organizations_by_district")
```
